```python
import numpy as np
from scipy.io import loadmat
import scipy.io
import scipy.signal as sig
import matplotlib.pyplot as plt
import pywt
```

```python
mat = scipy.io.loadmat('do_not_touch/Vowels/sub_4b_ch80_v_eog_removed_256Hz.mat')
eeg= mat['eeg_data_wrt_task_rep_no_eog_256Hz_last_beep']
idx=[*range(1,9),*range(10,32),*range(33,63)]
```

```python
scales = np.arange(1,129)
k=0
for i in range(100):
    for j in range(60):
        coef, freq = pywt.cwt(eeg[0,44+i][idx[j],:], scales, 'morl')
        a = np.resize(coef, (128,128))
        plt.imshow(abs(a), cmap='jet', aspect = 'auto', vmax= abs(a).max(), vmin = -abs(a).min())
        plt.axis('off')
        k+=1
        images_dir = 'images_sub4b/a_image'
        plt.savefig(f"{images_dir}/a_image_"+str(2640+k)+'.png', bbox_inches= 'tight', pad_inches=0)
```
